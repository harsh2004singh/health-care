```python
x = data.drop(['fetal_health'], axis=1)
y = data['fetal_health']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42,stratify = y)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_scaled = scaler.fit_transform(x)
```

# Random Forest

```python
rf = RandomForestClassifier( n_estimators=500,max_depth=10,min_samples_split=5,min_samples_leaf=2,random_state=42)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision (macro):", precision_score(y_test, y_pred_rf, average='macro'))
print("Recall (macro):", recall_score(y_test, y_pred_rf, average='macro'))
print("F1 Score (macro):", f1_score(y_test, y_pred_rf, average='macro'))

print("\nClassification Report")
print(classification_report(y_test, y_pred_rf))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred_rf))
```

# XGBoost

```python
xgb = XGBClassifier(n_estimators=300,max_depth=6,learning_rate=0.05,subsample=0.8,colsample_bytree=0.8,
                    objective='multi:softmax',num_class=3,random_state=42)
```

```python
# Shift labels from [1, 2, 3] → [0, 1, 2]
y_train_fixed = y_train - 1
y_test_fixed = y_test - 1
```

```python
# Train
xgb.fit(X_train, y_train_fixed)
```

```python
# Predict
y_pred_xgb = xgb.predict(X_test)
```

```python
# Map predictions back to original labels [1, 2, 3]
y_pred_xgb_original = y_pred_xgb + 1
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_xgb_original))
print("Precision (macro):", precision_score(y_test, y_pred_xgb_original, average='macro'))
print("Recall (macro):", recall_score(y_test, y_pred_xgb_original, average='macro'))
print("F1 Score (macro):", f1_score(y_test, y_pred_xgb_original, average='macro'))

print("\nClassification Report")
print(classification_report(y_test, y_pred_xgb_original))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred_xgb_original))
```
